```python
import pandas as pd
```

```python
#Reading the csv file
data=pd.read_csv('/kaggle/input/weather-data/Weather Data.csv')
```

```python
data
```

```python
#Show the first 10 rows in the data
data.head(10)
```

```python
#Provides basic information about the dataframe : row ,columns , types (if automatic detected)
data.info()
```

```python
#Find out different weather conditions available 
data['Weather'].unique()
```

```python
#Q: To find unique instances of weather types
weather=data['Weather'].value_counts()
dataweather=pd.DataFrame(weather)
dataweather = dataweather.reset_index()
dataweather.columns = ['Weather', 'Frequency'] # change column names
dataweather
```

```python
#Q: Find the number of times when the weather was exactly 'clear'
#data['Weather'].value_counts() it count occurrences of each unique value .
len(data[(data.Weather=='Clear')])
```

```python
#this is the another way to find the number of times when the weather was exactly 'clear'
data[data['Weather'] == 'Clear']
```

```python
#Q: To rename column named 'Weather' to Weather Condition'
data.rename(columns = {'Weather' : 'Weather_Condition'}, inplace=False)
```

```python
#Find all the unique wind speed value 
data['Wind Speed_km/h'].unique()
```

```python
#find the number of times when the wind speed was exactly 7km/h
len(data[(data['Wind Speed_km/h']==7)])
```

```python
#another way 
data['Wind Speed_km/h'].value_counts()[7]
```

```python
#Q: To find the mean temperature, wind speed and visibility
print(data['Temp_C'].mean())
print(data['Wind Speed_km/h'].mean())
print(data['Visibility_km'].mean())
```

```python
#Q: To find the days on which wind speed was less than or equal to 30 km/hr and temperature was greater than 0 C
windspeed=data[(data['Wind Speed_km/h'] <=30) & (data['Temp_C']>0)]
print(windspeed['Date/Time'])
print(windspeed.shape)
```

```python
#find out all the value of wind speed
data['Wind Speed_km/h'].value_counts()
```

```python
#Q: To find the date and temperatures for all instances when snow was recorded
snow=data[data['Weather'] == 'Snow']
snow.loc[:, ['Date/Time', 'Temp_C']]
```

```python
#Q: To display a pie chart of percentages of weather conditions

keep=dataweather[dataweather['Frequency']>20]
print(keep)
keep.plot.pie(y='Frequency',autopct='%1.1f%%', shadow=False, figsize=(12,9))
plt.title("Weather")
plt.show()
```
